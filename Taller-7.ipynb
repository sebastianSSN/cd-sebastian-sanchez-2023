```python
df["EmbarkedQ"] = 0
df.loc[df.Embarked == "Q", "EmbarkedQ"] = 1
df.loc[df.Embarked != "Q", "EmbarkedQ"] = 0

df["EmbarkedS"] = 0
df.loc[df.Embarked == "S", "EmbarkedS"] = 1
df.loc[df.Embarked != "S", "EmbarkedS"] = 0

df
```

```python
x =  df.iloc[:,[2,4,5,6,7,8,9,10,11]]
y =  df.Survived
logit_model=sm.Logit(y,x)
result=logit_model.fit()
print(result.summary())
```

```python
import numpy as np
from itertools import combinations

usefulColumns=[2,4,5,6,7,8,9,10,11]
combinaciones=[]

for i in range(1, len(usefulColumns)+1):
  temp = combinations(usefulColumns, i)
  for j in list(temp):
    combinaciones.append(np.array(j))
combinaciones
```

```python
#generar los pseudo r cuadrado
prsquared_values=[]

for i in range(len(combinaciones)):
  #Hacer regresion logistica
  x =  df.iloc[:, combinaciones[i]]
  y =  df.Survived
  print(combinaciones[i])
  logit_model=sm.Logit(y,x)
  result=logit_model.fit()
  prsquared_values.append(result.prsquared)
```

```python
combinacionesDF = pd.DataFrame()
combinacionesDF['combinaciones'] = combinaciones
combinacionesDF['pseudo_r_squared'] = prsquared_values
combinacionesDF
```

```python
#Ordenar las combinaciones de mejor a peor
bestCombinationsDF = combinacionesDF.sort_values(by='pseudo_r_squared', ascending=False)
bestCombinationsDF
```

```python
#Mostrar las 10 mejores combinaciones
bestCombinationsDF.head(10)
```

```python
#Definir cuantas se quieren mostrar
numComb=10
contador=0

for index, row in bestCombinationsDF.iterrows():
  #Mostrar numero de la combinacion
  contador=contador+1
  print("***Regresión logística número",contador,"***")
  print()
  print(row['combinaciones'])

  #Hacer regresion logistica
  x =  df.iloc[:, row['combinaciones']]
  y =  df.Survived
  logit_model=sm.Logit(y,x)
  result=logit_model.fit()
  print(result.summary())
  print()

  #gestionar cantidad de regresiones mostradas
  if contador == numComb:
        break
```

```python
#Hacer regresion logistica con mejor combinacion de 5
x =  df.iloc[:, mejor5]
y =  df.Survived
logit_model=sm.Logit(y,x)
result=logit_model.fit()
print(result.summary())
```

```python
from sklearn.model_selection import train_test_split

#Definir columnas a usar
x =  df.iloc[:, mejor5]
y =  df.Survived

# Separar train y test
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

#Guardar train y test como constante
X_train = sm.add_constant(X_train)
X_test = sm.add_constant(X_test)

#Entrenar modelo con train
logit_model = sm.Logit(y_train, X_train)
result = logit_model.fit()
print(result.summary())
```

```python
# Hacer predicciones con test y etiquetar como 1 o 0
y_prob = result.predict(X_test)
y_pred = (y_prob > 0.5).astype(int)
```

```python
#Generar matriz de confusion

from sklearn.metrics import confusion_matrix

confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```

```python
#obtener verdadero positivo, falso positivo, verdadero negativo y falso negativo
tn = confusion_matrix[0, 0]
fp = confusion_matrix[0, 1]
fn = confusion_matrix[1,0]
tp = confusion_matrix[1, 1]

print("Verdaderos negativos:",tn)
print("falsos positivos:",fp)
print("falsos negativos:",fn)
print("verdaderos positivos:",tp)
```

```python
#Obtener precision y exhaustividad

precision=tp/(tp+fp)
exhaustividad=tp/(tp+fn)

print("Precisión:",precision)
print("Exhaustividad:",exhaustividad)
```
