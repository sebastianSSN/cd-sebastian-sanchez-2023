```python
import pandas as pd
import matplotlib.pyplot as plt
import random

# Número de puntos que deseas generar
num_puntos = 1000000

# Genera puntos aleatorios en una cuadrícula de 1x1
puntos = [(random.uniform(0, 1), random.uniform(0, 1)) for _ in range(num_puntos)]

# Crea un DataFrame con los puntos
df1 = pd.DataFrame(puntos, columns=['X', 'Y'])

# Grafica los puntos
plt.figure(figsize=(10, 10))
# Disminuir el tamaño de los puntos
tamaño_puntos = 0.05

# Graficar los puntos con el tamaño especificado
plt.scatter(df1['X'], df1['Y'], s=tamaño_puntos)
plt.xlabel('Coordenada X', fontsize=15)
plt.ylabel('Coordenada Y', fontsize=15)
plt.title('Puntos aleatorios', fontsize=20)
plt.grid(True)
plt.show()
```

```python
import math

# Calcular el número de puntos en cada eje

puntos_totales = 1000000
puntos_lado=round(math.sqrt(puntos_totales))
distancia_entre_puntos=1/puntos_lado

# Crear una lista de coordenadas
coordenadas = []

# Generar puntos equidistantes en la cuadrícula de 1x1
for i in range(puntos_lado):
    for j in range(puntos_lado):
        x = i * distancia_entre_puntos
        y = j * distancia_entre_puntos
        coordenadas.append((x, y))

# Crear un DataFrame con las coordenadas
df2 = pd.DataFrame(coordenadas, columns=['X', 'Y'])

# Aumentar el tamaño de la gráfica
plt.figure(figsize=(8, 8))

# Grafica los puntos
plt.figure(figsize=(10, 10))
# Disminuir el tamaño de los puntos
tamaño_puntos = 0.05

# Graficar los puntos con el tamaño especificado
plt.scatter(df2['X'], df2['Y'], s=tamaño_puntos)
plt.xlabel('Coordenada X', fontsize=15)
plt.ylabel('Coordenada Y', fontsize=15)
plt.title('Puntos equisdistantes', fontsize=20)
plt.grid(True)
plt.show()
```

```python
#Función que calcula la distancia entre dos puntos
def distancia(x1,y1,x2,y2):
  d=math.sqrt(((x2-x1)**2)+((y2-y1)**2))
  return d

#función que cuenta los puntos con una distancia menor o igual a 1

def distanciaMenor(puntosDF):
  x = puntosDF["X"].values.tolist()
  y = puntosDF["Y"].values.tolist()
  largo = len(puntosDF)
  parejas = 0
  for i in range(0,largo):
    x1=x[i]
    y1=y[i]
    d=distancia(x1,y1,0,0)
    if (d<=1):
        parejas=parejas+1
  return parejas
```

```python

puntosMenores = distanciaMenor(df1)
print(puntosMenores)
piEstimado=(puntosMenores/puntos_totales)*4
print(piEstimado)
```

```python
puntosMenores = distanciaMenor(df2)
print(puntosMenores)
piEstimado=(puntosMenores/puntos_totales)*4
print(piEstimado)
```
