```python
#Regresión lineal múltiple
x = df0.iloc[:,[0,3,8]]
y = df0["Price"]
x2 = sm.add_constant(x)
est = sm.OLS(y, x2)
est2 = est.fit()
print(est2.summary())
```

```python
#r2
best_comb = [0,[0,0,0]]
randoms = [0,2,3,4,5,6,7,8]
#Regresión lineal múltiple

for i in randoms:
  for j in randoms:
    for k in randoms:
      if(i != j and i != k and j != k):
        x = df0.iloc[:,[i,j,k]]
        y = df0["Price"]
        x2 = sm.add_constant(x)
        est = sm.OLS(y, x2)
        est2 = est.fit()
        if((est2.rsquared)>best_comb[0]):
          best_comb[0] = est2.rsquared
          best_comb[1][0]= i
          best_comb[1][1] = j
          best_comb[1][2] = k

print(df0.columns[0])
print("Las mejor combinacion de columnas es: ",df0.columns[best_comb[1][0]]," - ",df0.columns[best_comb[1][1]]," - ",df0.columns[best_comb[1][2]])
print("Rsquared = ",best_comb[0])
```
